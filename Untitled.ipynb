```python
import seaborn as sns
from datetime import datetime

```

```python
def convert_to_date(row):
    
    if row[4] == '0000-00-00':
        return ""
    else:
        return datetime.strptime(row[4], '%Y-%m-%d')
```

```python
df["date_of_birth"] = df.apply(lambda x: convert_to_date(x), axis=1)
```
